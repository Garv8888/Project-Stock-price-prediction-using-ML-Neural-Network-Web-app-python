```python
Close_price = apple_data["Close"]
max(Close_price.values),min(Close_price.values)
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(Close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data=[]
y_data=[]
for i in range(100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
len(x_data)
```

```python
len(x_data)-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.7)

x_train=x_data[:splitting_len]
y_train=y_data[:splitting_len]
x_test=x_data[splitting_len:]
y_test=y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
```

```python
model = Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer= 'adam',loss = 'mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=1,epochs=2)
```

```python
model.summary()
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
inv_prediction = scaler.inverse_transform(prediction)
inv_prediction
```

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
rmse = np.sqrt(np.mean( (inv_prediction-inv_y_test)**2))
rmse
```

```python
rmse
```

```python
plotting_data=pd.DataFrame(
    {
        'original_test_data':inv_y_test.reshape(-1),
        'predictions' : inv_prediction.reshape(-1)
    },
    index=apple_data.index[splitting_len+100:]


)
plotting_data.head()
```

```python
plotgraph((15,6),plotting_data,"test data")
```

```python
plotgraph((15,6),pd.concat([Close_price[:splitting_len+100],plotting_data],axis=0),"whole data")
```
